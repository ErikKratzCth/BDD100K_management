```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean

data_main_folder = Path('C:/Users/erikk/Dataset exjobb/BDD100K')
images_folder = data_main_folder / "bdd100k_images/bdd100k/images/100k/train_and_val"
labels_folder = data_main_folder / "bdd100k_labels_release/bdd100k/labels"
labels_file = labels_folder / "bdd100k_labels_images_train_and_val.json"

with open(labels_file) as json_data:
    labels = json.load(json_data) # json.load() creates a list with one dictionary for every image
```

```python
# Distribution of number of different objects across image set
import numpy as np

target = "car"
car_counter = np.zeros(len(labels))
for idx, image in enumerate(labels):
    cars = 0;
    for object in image["labels"]:
        if object["category"] == target:
            cars += 1;
    car_counter[idx] = cars

import matplotlib.pyplot as plt

plt.hist(car_counter, bins='auto')  # arguments are passed to np.histogram
plt.title("Number of "+ target + "s in image set")

plt.show()
```

```python
# Distribution of total number of objects across image set
import numpy as np

object_counter = np.zeros(len(labels))
for idx, image in enumerate(labels):
    object_counter[idx] = len(image["labels"])

import matplotlib.pyplot as plt

plt.hist(object_counter, bins='auto')  # arguments are passed to np.histogram
plt.title("Total number of objects per image in validation set")
plt.show()
```

```python
import tensorflow as tf

# extract attributes from labels dictionaries and convert to one hot labels
weather = list()
for entry in labels:
    weather.append(entry["attributes"]["weather"])
weather_types = list(set(weather))
print(weather_types)
for i in range(0,len(weather_types)):
    weather = [i if el == weather_types[i] else el for el in weather]

time_of_day = list()
for entry in labels:
    time_of_day.append(entry["attributes"]["timeofday"])
time_types = list(set(time_of_day))
print(time_types)
for i in range(0,len(time_types)):
    time_of_day = [i if el == time_types[i] else el for el in time_of_day]
    
scene = list()
for entry in labels:
    scene.append(entry["attributes"]["scene"])
scene_types = list(set(scene))
print(scene_types)
for i in range(0,len(scene_types)):
    scene = [i if el == scene_types[i] else el for el in scene]
```

```python
import numpy as np
category_counter = np.zeros([4,7,7])

# count number of labels in each weather category joint with timeofday category
for idx in range(0,len(labels)):
    category_counter[time_of_day[idx]][weather[idx]][scene[idx]] += 1

threshold = 0
n_valid_images = 0

print("All well defined categories (no attribute is 'undefined') with more than " + str(threshold) + " images:\n")
for idx in range(0,len(time_types)):
    if time_types[idx] != "undefined":
        for idx_2 in range(0,len(weather_types)):
            if weather_types[idx_2] != "undefined":
                for idx_3 in range(0,len(scene_types)):
                    if scene_types[idx_3] != "undefined" and category_counter[idx][idx_2][idx_3] > threshold:
                        n_valid_images += category_counter[idx][idx_2][idx_3]
                        print(time_types[idx] + "\t\t" + weather_types[idx_2] + "\t\t" + scene_types[idx_3] + ":\t" + str(category_counter[idx][idx_2][idx_3]))
print("In total there are " + str(n_valid_images) + " images with no undefined cateogory\n")
#print(category_counter)
#print(sum(sum(category_counter)))
print(scene_types)
```

```python
import numpy as np
weather_type_counter = np.zeros(7) # in order clear, overcast, rainy, snowy, undefined

# count number of labels in each weather category
for entry in labels:
    weather = entry["attributes"]["weather"]
    if weather == "clear":
        weather_type_counter[0] += 1
    elif weather == "overcast":
        weather_type_counter[1] += 1
    elif weather == "rainy":
        weather_type_counter[2] += 1
    elif weather == "snowy":
        weather_type_counter[3] += 1
    elif weather == "undefined":
        weather_type_counter[4] += 1
    elif weather == "partly cloudy":
        weather_type_counter[5] += 1
    elif weather == "foggy":
        weather_type_counter[6] += 1
   
print(weather_type_counter)
print(sum(weather_type_counter))

time_type_counter = np.zeros(4) # in order clear, overcast, rainy, snowy, undefined

# count number of labels in each weather category
for entry in labels:
    time_var = entry["attributes"]["timeofday"]
    if time_var == "undefined":
        time_type_counter[0] += 1
    elif time_var == "daytime":
        time_type_counter[1] += 1
    elif time_var == "night":
        time_type_counter[2] += 1
    elif time_var == "dawn/dusk":
        time_type_counter[3] += 1
   
print(time_type_counter)
print(sum(time_type_counter))
```
